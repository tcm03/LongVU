### This is an analysis of the filtered EnTube dataset with all the inaccessible links removed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EnTube_filtered.csv')
df.head()
```

```python
df.info()
```

```python
category_count = df['category'].value_counts()
category_count.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.title('Number of Videos per Category')
plt.show()
```

```python
engagement_count = df['engagement_rate_label'].value_counts()
engagement_count.plot(kind='bar')
plt.xlabel('Engagement Rate')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.title('Number of Videos per Engagement Rate')
plt.show()
```

```python
all_labels = set(df['engagement_rate_label'])
for label in all_labels:
    min_qscore = df[df['engagement_rate_label'] == label]['q_score'].min()
    max_qscore = df[df['engagement_rate_label'] == label]['q_score'].max()
    print(f'{label}: min_qscore={min_qscore}, max_qscore={max_qscore}')
```
